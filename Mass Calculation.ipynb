```python
import numpy as np

# Constants
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
AU_to_meters = 1.496e+11  # Conversion factor from AU to meters
years_to_seconds = 3.156e+7  # Conversion factor from years to seconds
mass_earth = 5.972e+24  # Mass of Earth in kg

# Given values for Saturn
a_AU_saturn = 9.5  # Semi-major axis in AU
T_years_saturn = 29.4  # Orbital period in years

# Convert semimajor axis from AU to meters
a_meters_saturn = a_AU_saturn * AU_to_meters
print(f"Semi-major axis of Saturn converted to meters: {a_meters_saturn} m")

# Convert orbital period from years to seconds
T_seconds_saturn = T_years_saturn * years_to_seconds
print(f"Orbital period of Saturn converted to seconds: {T_seconds_saturn} s")

# Calculate mass of Saturn using the simplified Kepler's third law
M_saturn = (4 * np.pi**2 * a_meters_saturn**3) / (G * T_seconds_saturn**2)

# Convert mass of Saturn from kg to Earth masses for easier interpretation
M_saturn_in_earth_masses = M_saturn / mass_earth

print(f"Mass of Saturn: {M_saturn} kg")
print(f"Mass of Saturn in Earth masses: {M_saturn_in_earth_masses:.2f}")
```
